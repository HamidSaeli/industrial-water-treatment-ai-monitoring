```python
import numpy as np
import pandas as pd


pd.set_option("display.max_columns", None)

df = pd.read_csv("../data/raw/merged.csv")
```

```python
df.columns = df.columns.str.strip()

df.columns.tolist()
```

```python
duplicate_rows = df.duplicated().sum()

print(f"Duplicate Rows: {duplicate_rows}")
```

```python
df["Timestamp"] = (
    df["Timestamp"]
        .astype(str)
        .str.strip()
)

df["Timestamp"] = pd.to_datetime(
    df["Timestamp"],
    format="mixed",
    dayfirst=True
)
```

```python
df["Timestamp"].min(), df["Timestamp"].max()
```

```python
df["Normal/Attack"].value_counts()
```

```python
df["Normal/Attack"].value_counts(normalize=True) * 100
```

```python
missing = (
    df.isnull()
      .sum()
      .sort_values(ascending=False)
)

missing.head(20)
```

```python
missing_percent = (
    df.isnull().mean() * 100
).sort_values(ascending=False)

missing_percent.head(20)
```

```python
constant_columns = []

for col in df.columns:
    if df[col].nunique() == 1:
        constant_columns.append(col)

constant_columns
```

```python
numeric_columns = df.select_dtypes(include=np.number).columns

len(numeric_columns)
```

```python
categorical_columns = df.select_dtypes(exclude=np.number).columns

categorical_columns
```

```python
df.to_csv("../data/processed/swat_stage1_clean.csv", index=False)
```
